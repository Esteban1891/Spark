**Instalacion de JAVA**

1. Actualizar CLI: `sudo apt update -y`
2. Instalar Java: `sudo apt install default-jre`
3. Instalar JRE: `sudo apt install openjdk-11-jre`
4. Instalar jdk: `sudo apt install openjdk-11-jdk `
5. Java Instalado: `java --version`

**Instalacion de Spark**

1. Ingresar superuser: `sudo su` y luego `mkdir –p /opt/spark`
2. Descargar Spark: `sudo apt install wget` y luego `wget  https://dlcdn.apache.org/spark/spark-3.2.1/spark-3.2.1-bin-hadoop3.2.tgz`
3. Descomprimir Spark: `tar xvf spark-3.2.1-bin-hadoop3.2.tgz`
4. Con el super usuario salimos de la carpeta `/opt/spark` con el comando `cd`, luego escribimos `ls -al`, donde veremos todos los archivos que estan ocultos, Alli vemos uno que se llama `.bashrc`

5. Añadir Entornos: Dentro de vim vamos a ejecutar `nano .bashrc` y despues de abierto, con la fecha abajo nos dirigimos al final:
    - Copiamos `export SPARK_HOME=/opt/spark/spark-3.2.1-bin-hadoop3.2` y lo pegamos
    - Copiamos `export PATH=$PATH:$SPARK_HOME/bin` y lo pegamos

    Luego guardamos con `Ctrl + X`, escribimos `Y` y finalmente damos `enter`

    Podemos validar si escribimos bien, escribimos `cat .bashrc`

6. Escribimos `source .bashrc`
7. Para validar la instalacion de Spark escribimos `spark-shell`, y alli se ve la version de Spark instalada correctamente =)


**Instalacion de Python**

1. Instalacion Python3: `sudo apt install python3`

**Instalacion de PIP**

1. Instalacion PIP: `sudo apt install python3-pip`
2. Añadir entornos: Dentro de vim vamos a ejecutar `nano .bashrc` y despues de abierto, con la fecha abajo nos dirigimos al final:
    - Copiamos `export PYSPARK_PYTHON=/usr/bin/python3` y lo pegamos
    
    Luego guardamos con `Ctrl + X`, escribimos `Y` y finalmente damos `enter`

    Podemos validar si escribimos bien, escribimos `cat .bashrc`

**Instalacion de Pyspark**

1. Instalacion PySpark: `sudo pip install pyspark`
2. Validacion de Instalacion: Primero ejecutamos `source .bashrc` y luego escribimos `pyspark`, debe aparecer la misma version de Spark previamente instalada =)
